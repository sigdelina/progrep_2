```python
import json
import urllib.request
#так как в задании было сказано "Для каждой из перечисленных операций нужно написать свою функцию", то функций всего 5 + main()
#поэтому каждая из функций написана полноценно и (почти - кроме yaziki(data)) обосблена от дргуих

def finding_Rep(sp): # функция запрашивает пользователя и выводит его репозитории с описанием
    stroka = '' # строка для пользователей списка
    for nam in sp: # перебираем пользоваелей и добавляем их имена в строку
        if nam != sp[-1]:
            stroka += nam + ', '
        elif nam == sp[-1]:
            stroka += nam + '.'
    print('Вам предлагается на выбор список пользователей: ', stroka)
    user = input('Выбирите пользователя из списка: ')
    while user not in sp: # цикл для правильного ввода имени пользователя
        user = input('Введенного Вами имени пользователя нет в списке. Попробуйте снова)')
        if user in sp:
            break
    print("\n***\nВы выбрали пользователя", user, '.\n')
    token = "5e9bfb42367d1c5dd4f558711703d2c2776baaed" # токен из гитхаба
    url = 'https://api.github.com/users/%s/repos?page=1?access_token=%s&per_page=100' % (user, token) # ссылка
    response = urllib.request.urlopen(url)  # запрос серверу
    text = response.read().decode('utf-8')  # перевод ответа от запроса в строку и изменение кодировки
    data = json.loads(text) # переводим json в объект питона - объекты = репозитории
                        #print(len(data))    # можно распечатать, сколько у пользователя репозиториев
    print("Список репозиториев пользователя", user, "с описанием:\n")
    for rp in data: # из полученного ответа выбираем названия репозиториев и их описания и выводим
        if rp != data[-1]:
            print(rp["name"], ': ', rp["description"], ',') # и распечатать названия всех репозиториев
        elif rp == data[-1]:
            print(rp["name"], ': ', rp["description"])
    return data
```

```python
def yaziki(data): #функция просматривает репозитории и ищет языки: выводим их и считает количесвто
    new_s = '' #текстовая строка, в которую ляжет предложение с языками
    spis = [] #список
    lingua = {} #частотоный словарь
    for i in data: #для элементов в информации с страницы гитхаба: ищутся языки из репозиториев и заносятся в частотный словарь
        yaz = i["language"]
        if yaz in lingua:
            lingua[yaz] += 1
        else:
            lingua[yaz] = 1 
    #print(lingua)
    del lingua[None] #удаляем None как язык, потому что значит, что у реп. не укзаан язык
    #print(lingua)
    for lin in lingua: #переносим элементы словаря в элементы списка
        spis.append(lin)
    #print(spis)
    for lang in spis: #пополняем строку элементами списка - чтобы красиво вывелось
        if lang != spis[-1]:
            new_s += str(lang) + ', '
        elif lang == spis[-1]:
            new_s += str(lang)
    #print(new_s)
    print('\nПользователь пишет на', new_s, '.')
    print('\nЯзыки используются в следующем количестве его репозиториев:') 
    por = sorted(lingua.items(), key=lambda x: x[1], reverse=True) #список кортежей: cортируем словарь по убыванию
    ling_2 = dict(por) #из списка снова делаем словарь с нужной последовательностью
    for l in ling_2: # красивый вывод
        print(l, " --> ", ling_2[l])
    return ling_2
```

```python
def allof(sp): #функция просматривает список и счаитает максимальное количество репозиториев
    stro_3 = '' #строка для красивой записи
    d_name = {} #словарь для пользователей и количества их репозиториев
    for user in sp: #перебираем пользователей из списка 
        token = "5e9bfb42367d1c5dd4f558711703d2c2776baaed" #токен из гитхаба
        url = 'https://api.github.com/users/%s/repos?page=1?access_token=%s&per_page=100' % (user, token) #ссылка
        response = urllib.request.urlopen(url) #запрос серверу
        text = response.read().decode('utf-8')  #перевод ответа от запроса в строку и изменение кодировки
        data = json.loads(text) #переводим json в объект питона - объекты = репозитории
        #print(len(data))
        d_name.update({user: len(data)}) #пополняем словарь пользователями и количеством их репозитриев
    #print('\n', d_name)
    vib = sorted(d_name.items(), key=lambda x: x[1], reverse=True) #список сортированных кортежей (из словаря) по убыванию
    #print(vib)
    new = dict(vib) #перевод списка кортежей в словарь с значениями по убыванию
    #print(new)
    maxv = max(new.values()) #определение максимальных значений в словаре
    choice = [key for key, value in new.items() if value == maxv] #список пользователей(ключей), где значения из словаря макисмальны
    #print(choice)
    for el in choice: #перебор элментов из списка и добавяление нужных пользователей в строку
        if el != choice[-1]:
            stro_3 += el + ', '
        elif el == choice[-1]:
            stro_3 += el 
    print('\n***\nИз всего списка у', stro_3, 'больше всего репозиториев.')
    print('P.S. максимальное число репозиториев =', maxv)
    return stro_3
```

```python
def kolich_Yaz(sp): #функция ищет самый распространенный язык у пользователей
    stro_4 = '' #строка для красивого вывода
    koll = {} #частотный словарь для количества языков
    for user in sp: #перебираем пользователей из списка
        token = "5e9bfb42367d1c5dd4f558711703d2c2776baaed" #токен из гитхаба
        url = 'https://api.github.com/users/%s/repos?page=1?access_token=%s&per_page=100' % (user, token) #ссылка
        response = urllib.request.urlopen(url) #запрос серверу
        text = response.read().decode('utf-8') #перевод ответа от запроса в строку и изменение кодировки
        data = json.loads(text) #переводим json в объект питона - объекты = репозитории
        for lan in data: #в репозитории пользователя ищем язык, на котором он пишет, пополнаяем словарь
            l = lan["language"]
            if l in koll:
                koll[l] += 1
            else:
                koll[l] = 1
    #print(koll)
    del koll[None] #избавляемся от случаев, когда у пользователя не указан язык репозитория
    #print(koll)
    sp_ko = sorted(koll.items(), key=lambda x: x[1], reverse=True) #список кортежей: cортируем словарь по убыванию
    #print(sp_ko)
    fin = dict(sp_ko) #из списка снова делаем словарь с нужной последовательностью
    #print(fin)
    maxl = max(fin.values()) #максимальное значение в словаре
    choice_2 = [key for key, value in fin.items() if value == maxl] #список языков(ключей), где значения из словаря макисмальны
    #print(choice_2)
    if len(choice_2) >= 1: #для списка, в котором есть хотя бы 1 элемент
        for la in choice_2: #перебиаем языки в списке и добавляем их в строку
            if la != choice_2[-1]:
                stro_4 += la + ', '
            elif la == choice_2[-1]:
                stro_4 += la 
        if len(choice_2) == 1: #выводим данные, если самый частый язык один
            print('\n***\nСамый частый язык: ', stro_4)
        elif len(choice_2) > 1: #выводим данные, если самых частых языков несколько
            print('\n***\nСамые частые языки: ', stro_4)
    else:
        print('Языки репозиториев не указаны ни у одного пользователя')
    return stro_4
```

```python
def followers(sp): #функция, считающая подписчиков
    str_4 = ''  #строка для удобного вывода
    dict_fol = {} #словарь для пользователей и количества подписчиков
    for user in sp: #перебор пользователей из списка
        token = "5e9bfb42367d1c5dd4f558711703d2c2776baaed" #токен из гитхаба 
        url_1 = 'https://api.github.com/users/%s/followers?page=1?access_token=%s&per_page=100' % (user, token) #ссылка
        response_1 = urllib.request.urlopen(url_1) #запрос серверу
        text_1 = response_1.read().decode('utf-8') #перевод ответа от запроса в строку и изменение кодировки
        data_1 = json.loads(text_1) #переводим json в объект питона - объекты = фолловеры
        dict_fol.update({user: len(data_1)}) #пополение словаря пользователями и количесвтом подписчков
    #print(dict_fol)
    flw = sorted(dict_fol.items(), key=lambda x: x[1], reverse=True) #список сортированных кортежей по убыванию (по словарю)
    #print(flw)
    nf = dict(flw) #переврдим список кортежей в словарь
    #print(nf)
    maxfl = max(nf.values()) #максимальное значение в словаре
    choice_3 = [key for key, value in nf.items() if value == maxfl] #список для пользователей(ключ) с максимальными значениями
    #print(choice_3)
    for fow in choice_3: #добавляем элементы списка в строку
        if fow != choice_3[-1]:
            str_4 += fow + ', '
        elif fow == choice_3[-1]:
            str_4 += fow 
    print('\n***\nИз всего списка у', str_4, 'больше всего фоловверов.')
    print('P.S. максимальное число подписчиков =', maxfl)
    return str_4
```
