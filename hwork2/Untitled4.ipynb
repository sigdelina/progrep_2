```python
import json
import urllib.request
# так как в задании было сказано
# "Для каждой из перечисленных операций нужно написать свою функцию"
# то функций всего 5 + main()
# поэтому каждая из функций написана полноценно,почти обосбленно от дргуих
# Чтобы сохранить обособленность функция -
# просмотр страниц везде одинаковый и сделан циклом :
# (без понятия, как это уменьшить - поэтому функции большие из-за цикла)
```

```python
# функция запрашивает пользователя и выводит его репозитории с описанием
def finding_Rep(sp, token):
    stroka = ''  # строка для пользователей списка
    for nam in sp:  # перебираем пользоваелей и добавляем их имена в строку
        if nam != sp[-1]:
            stroka += nam + ', '
        elif nam == sp[-1]:
            stroka += nam + '.'
    print('Вам предлагается на выбор список пользователей: ', stroka)
    user = input('Выбирите пользователя из списка: ')
    while user not in sp:  # цикл для правильного ввода имени пользователя
        user = input('Введенного Вами имени пользователя нет в списке.'
                     'Попробуйте снова)')
        if user in sp:
            break
    print("\n***\nВы выбрали пользователя", user, '.\n')
    page = 1
    url = \
    'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
        % (user, page, token)
    response = urllib.request.urlopen(url)
    text = response.read().decode('utf-8')  # читаем ответ в строку
    data = json.loads(text)  # превращаем джейсон-строку в объекты питона
    while len(data) % 100 == 0:  # делаем цикл: если на странице 100 элементов
        page += 1
        url = \
        'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
            % (user, page, token)
        response = urllib.request.urlopen(url)
        text = response.read().decode('utf-8')  # читаем ответ в строку
        data += json.loads(text)  # превращаем джейсон-строку в объекты питона
        if len(data) % 100 != 0:
            break
        continue
    print("Список репозиториев пользователя", user, "с описанием:\n")
    for rp in data:  # из ответа -> выбор названия репозиториев и их описания
        if rp != data[-1]:
            print(rp["name"], ': ', rp["description"], ',')
        elif rp == data[-1]:
            print(rp["name"], ': ', rp["description"])
    return data
```

```python
# функция просматривает репозитории и ищет языки: вывод и количесвто
def yaziki(data):
    new_s = ''  # текстовая строка, в которую ляжет предложение с языками
    spis = []  # список
    lingua = {}  # частотоный словарь
    for i in data:  # языки из репозиториев и заносятся в частотный словарь
        yaz = i["language"]
        if yaz in lingua:
            lingua[yaz] += 1
        else:
            lingua[yaz] = 1
    if None in lingua:
        del lingua[None]  # удаляем None как язык
    for lin in lingua:  # переносим элементы словаря в элементы списка
        spis.append(lin)
    for lang in spis:  # пополняем строку элементами списка
        if lang != spis[-1]:
            new_s += str(lang) + ', '
        elif lang == spis[-1]:
            new_s += str(lang)
    print('\nПользователь пишет на', new_s, '.')
    print('\nЯзыки используются в следующем количестве его репозиториев:')
    # список кортежей: cортируем словарь по убыванию
    por = sorted(lingua.items(), key=lambda x: x[1], reverse=True)
    ling_2 = dict(por)  # из списка делаем словарь с нужной последовательностью
    for l in ling_2:  # красивый вывод
        print(l, " --> ", ling_2[l])
    return ling_2
```

```python
# функция просматривает список и счаитает максимальное количество репозиториев
def allof(sp, token):
    stro_3 = ''  # строка для красивой записи
    d_name = {}  # словарь для пользователей и количества их репозиториев
    for user in sp:  # перебираем пользователей из списка
        page = 1
        url = \
        'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
            % (user, page, token)
        response = urllib.request.urlopen(url)
        text = response.read().decode('utf-8')  # читаем ответ в строку
        data = json.loads(text)  # превращаем джейсон-строку в объекты питона
        while len(data) % 100 == 0:  # делаем цикл: на странице 100 эле-ов 
            page += 1
            url = \
            'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
                % (user, page, token)
            response = urllib.request.urlopen(url)
            text = response.read().decode('utf-8')  # читаем ответ в строку
            data += json.loads(text)  # джейсон-строку в объекты питона
            if len(data) % 100 != 0:
                break
            continue
        d_name.update({user: len(data)})  # слов. пользователей и репозитриев
    # список сортированных кортежей (из словаря) по убыванию
    vib = sorted(d_name.items(), key=lambda x: x[1], reverse=True)
    new = dict(vib)  # перевод списка кортежей в словар по убыванию
    maxv = max(new.values())  # определение максимальных значений в словаре
    # список пользователей(ключей), где значения из словаря макисмальны
    choice = [key for key, value in new.items() if value == maxv]
    for el in choice:  # перебор элментов из списка и добавяление в строку
        if el != choice[-1]:
            stro_3 += el + ', '
        elif el == choice[-1]:
            stro_3 += el
    print('\n***\nИз всего списка у', stro_3, 'больше всего репозиториев.')
    print('P.S. максимальное число репозиториев =', maxv)
    return stro_3
```

```python
# функция ищет самый распространенный язык у пользователей
def kolich_Yaz(sp, token):
    stro_4 = ''  # строка для красивого вывода
    koll = {}  # частотный словарь для количества языков
    for user in sp:  # перебираем пользователей из списка
        page = 1
        url = \
        'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
            % (user, page, token)
        response = urllib.request.urlopen(url)
        text = response.read().decode('utf-8')  # читаем ответ в строку
        data = json.loads(text)  # превращаем джейсон-строку в объекты питона
        while len(data) % 100 == 0:  # делаем цикл: если на странице 100 элементов
            page += 1
            url = \
            'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
                % (user, page, token)
            response = urllib.request.urlopen(url)
            text = response.read().decode('utf-8')  # читаем ответ в строку
            data += json.loads(text)  # джейсон-строку в объекты питона
            if len(data) % 100 != 0:
                break
            continue
        for lan in data:  # язык репозитория, пополнаяем словарь
            l = lan["language"]
            if l in koll:
                koll[l] += 1
            else:
                koll[l] = 1
    if None in koll:
        del koll[None]  # когда у пользователя не указан язык репозитория
    # список кортежей: cортируем словарь по убыванию
    sp_ko = sorted(koll.items(), key=lambda x: x[1], reverse=True)
    fin = dict(sp_ko)  # из списка делаем словарь с нужной последовательностью
    maxl = max(fin.values())  # максимальное значение в словаре
    # список языков(ключей), где значения из словаря макисмальны
    choice_2 = [key for key, value in fin.items() if value == maxl]
    if len(choice_2) >= 1:  # для списка, в котором есть хотя бы 1 элемент
        for la in choice_2:  # перебиаем языки в списке и добавляем их в строку
            if la != choice_2[-1]:
                stro_4 += la + ', '
            elif la == choice_2[-1]:
                stro_4 += la
        if len(choice_2) == 1:  # выводим данные, если самый частый язык один
            print('\n***\nСамый частый язык: ', stro_4)
        elif len(choice_2) > 1:  # данные, если самых частых языков несколько
            print('\n***\nСамые частые языки: ', stro_4)
    else:
        print('Языки репозиториев не указаны ни у одного пользователя')
    return stro_4
```

```python
# функция, считающая подписчиков
def followers(sp, token):
    str_4 = ''  # строка для удобного вывода
    dict_fol = {}  # словарь для пользователей и количества подписчиков
    for user in sp:  # перебор пользователей из списка
        page = 1
        url_1 =\
        'https://api.github.com/users/%s/followers?page=%s&per_page=100&access_token=%s'\
            % (user, page, token)
        response_1 = urllib.request.urlopen(url_1)  # запрос серверу
        text_1 = response_1.read().decode('utf-8')  # перевод ответа в строку
        data_1 = json.loads(text_1)  # json в объект питона-объекты = фолловеры
        while len(data_1) % 100 == 0:  # цикл: кол-во делится на 100 без остатка
            page += 1
            url_1 = \
            'https://api.github.com/users/%s/repos?page=%s&per_page=100&access_token=%s'\
                % (user, page, token)
            response_1 = urllib.request.urlopen(url_1)
            text_1 = response_1.read().decode('utf-8')  # читаем ответ в строку
            data_1 += json.loads(text_1)  # джейсон-строку в объекты питона
            if len(data_1) % 100 != 0:
                break
            continue
        dict_fol.update({user: len(data_1)})  # словарь для кол. подписчков
    # список сортированных кортежей по убыванию (по словарю)
    flw = sorted(dict_fol.items(), key=lambda x: x[1], reverse=True)
    nf = dict(flw)  # переврдим список кортежей в словарь
    maxfl = max(nf.values())  # максимальное значение в словаре
    # список для пользователей(ключ) с максимальными значениями
    choice_3 = [key for key, value in nf.items() if value == maxfl]
    for fow in choice_3:  # добавляем элементы списка в строку
        if fow != choice_3[-1]:
            str_4 += fow + ', '
        elif fow == choice_3[-1]:
            str_4 += fow
    print('\n***\nИз всего списка у', str_4, 'больше всего фоловверов.')
    print('P.S. максимальное число подписчиков =', maxfl)
    return str_4
```
